```
import os
import numpy as np
import cv2

X = np.zeros((1264,64,64,3))
Y = np.zeros((1264,10))

# 2528 = 63284
#Y = [left,right,straight,up,angry,happy,neutral,sad,open,sunglasses]
```

```
count = 0
image_dir = "/content/drive/My Drive/Advanced_Assignment_Dataset"
for root,dirs,files in os.walk(image_dir):
  for file in files:
    path = os.path.join(image_dir,file)
    character = file.split("_")
    if character[0] == 'left':
      Y[count,0] = 1
    if character[0] == 'right':
      Y[count,1] = 1
    if character[0] == 'straight':
      Y[count,2] = 1
    if character[0] == 'up':
      Y[count,3] = 1
    if character[1] == 'angry':
      Y[count,4] = 1
    if character[1] == 'happy':
      Y[count,5] = 1
    if character[1] == 'neutral':
      Y[count,6] = 1
    if character[1] == 'sad':
      Y[count,7] = 1
    if character[2] == 'open':
      Y[count,8] = 1
    if character[2] == 'sunglasses':
      Y[count,9] = 1
    Y_all = Y[count,:]

    img = cv2.imread(path,1)

    img = cv2.resize(img,(64,64))
    rows,cols,_ = img.shape
    X[count,:,:] = img
    Y[count,:] = Y_all
    count = count + 1

    cropped = img[10:54,10:54]
    cropped = cv2.resize(cropped,(64,64))
    img1 = cropped
    X[count,:,:] = img1
    Y[count,:] = Y_all
    count = count + 1

    print(count - 1)
```

```
X = X.reshape((1264,64,64,3))
```

```
from sklearn.utils import shuffle
X, Y = shuffle(X, Y)
cv2_imshow(X[0,:,:])
cv2_imshow(X[1,:,:])
cv2_imshow(X[2,:,:])
cv2_imshow(X[3,:,:])
cv2_imshow(X[4,:,:])
cv2_imshow(X[5,:,:])
cv2_imshow(X[6,:,:])
print(Y[0])
print(Y[1])
print(Y[2])
print(Y[3])
print(Y[4])
print(Y[5])
print(Y[6])
X = X/255
print(X[1])
```

```
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.2)
print(Xtrain.shape)
print(Xtest.shape)
print(Ytest.shape)
print(Ytrain.shape)
```

```
import tensorflow as tf
```

```
model = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=(64,64,3)),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D(pool_size=(2,2),),
        tf.keras.layers.Dropout(0.5),

        tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D(pool_size=(2,2),),
        tf.keras.layers.Dropout(0.5),

        tf.keras.layers.Conv2D(128,(3,3),activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D(pool_size=(2,2),),
        tf.keras.layers.Dropout(0.5),
        
        tf.keras.layers.Flatten(),

        tf.keras.layers.Dense(256,activation='relu'),
        tf.keras.layers.BatchNormalization(),

        tf.keras.layers.Dense(10, activation = 'sigmoid')
])
```

```
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])
```

```
model.summary()
```

```
epochs=80
history = model.fit(Xtrain,Ytrain,batch_size=128,epochs=epochs,validation_data=(Xtest,Ytest))
model.save_weights("weights4.h5")
```

```
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
